```python
# Assign Initial value

S0=50.0       # spot stock price
K=52.0        # strike price
t=2.0         # maturity 2 years 
r=0.05        # risk free interest rate 5%
sigma=0.3     # volatility
```

```python
#Pricing American Option By using binomial tree

N = 2                                                               # number of time steps  
payoff = "put"                                                      # option Type 

dT = float(t) / N                                                   # Delta t
u = np.exp(sigma * np.sqrt(dT))                                     # Probability to increase price
d = 1.0 / u                                                         # Probability to decrease price 

V = np.zeros(N+1)                                                   # initialize the price vector
S_T = np.array( [(S0 * u**j * d**(N - j)) for j in range(N + 1)] )  # price S_T at time T

a = np.exp(r * dT)                                                  # risk free compound return
p = (a - d)/ (u - d)                                                # risk neutral up probability
q = 1.0 - p                                                         # risk neutral down probability   

# Payoff for call and put
if payoff =="call":
    V[:] = np.maximum(S_T-K, 0.0)
elif payoff =="put":
    V[:] = np.maximum(K-S_T, 0.0)

for i in range(N-1, -1, -1):
    V[:-1] = np.exp(-r*dT) * (p * V[1:] + q * V[:-1])               # the price vector is overwritten at each step
    S_T = S_T * u                                                   # way to obtain the price at the previous time step
    if payoff=="call":
        V = np.maximum( V, S_T-K )
    elif payoff=="put":
        V = np.maximum( V, K-S_T )
    
print("American Option Price by using Binomial Tree: €", np.round (V[0], 2))
```

## American put option exercise boundary

```python
# Import libraries
from scipy.stats import norm
from math import log, sqrt,exp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Assign required value for MC
T = 3/12                                                             # Maturity 3 Months in a year
n = 1000                                                             # Number of paths
m = 4                                                                # Number of Time step
```

#### Use Monte Carlo Simulation for m time steps and n different paths. It will Return a matrix of stock prices with each row as a path and each column as a time increment.

```python
#MC price Simulation

def price_simulations(S0, r, sigma, T, m, n):

    dt = T/m                                                          # Length of time increment (years) (Delta)
    
    total_paths = []
    for i in range(n):
        path_i = [S0]
        for j in range(m):
            stj = path_i[-1]*np.exp((r-sigma**2/2)*dt + sigma*np.sqrt(dt)*np.random.normal(0,1))
            path_i.append(stj)
        total_paths.append(path_i)
    return total_paths
```

```python
# Visualize all path of Monte Carlo Simulation
total_paths = price_simulations(S0, r, sigma, T, m, n)

# For Graph
plt.figure(figsize = (5, 5))
for i in range(n):
    ax = sns.lineplot(x = range(m+1), y = total_paths[i])
plt.title('Monte Carlo Simulation\nAll Paths', fontsize = 20)
plt.xlabel('Time Step', fontsize = 16)
plt.ylabel('Stock Price (€)', fontsize = 16)
ax.xaxis.set_tick_params(labelsize = 16)
ax.yaxis.set_tick_params(labelsize = 16)
```

```python
# function For Optiomal Exerise boundry

def opt_exer_bounds(S, P, K, n, thetaj_p1):

# S = Stock price for all paths at a given time.
# P = Put price for all paths at a given time assuming all paths will exercise in the future and compare with exercise today
# in this function.
# thetaj_p1 = Optimal Exercise Boundary for the next time step
 
    SP_pairs = [list(pair) for pair in zip(S, P)]
    previous_P = []

    put_hat = np.mean(P)
    thetaj_hat = thetaj_p1
    for i in range(n):
        if SP_pairs[i][0] <= thetaj_p1:
            put = []
            for j in range(n):
                if SP_pairs[j][0] <= SP_pairs[i][0]:
                    put.append(max(0, K-SP_pairs[j][0]))
                else:
                    put.append(SP_pairs[j][1])
            if np.mean(put) > put_hat:
                put_hat = np.mean(put)
                thetaj_hat = SP_pairs[i][0]
                previous_P = put
    if previous_P == []:
        previous_P = P

    return (previous_P, thetaj_hat)
```

####    For Optimal Exercise Boundary backwards induction is used to determine optimal put pricing at a given time.

```python
#Function for creating Opion Price & exercise boundary

def MC_American_put(r, sigma, S0, K, T, m, n):

    # MC Simulation
    total_paths = price_simulations(S0, r, sigma, T, m, n)
    dt = T/m                                                                            #(Delta)
    
    # Price & optimal exercise point At maturity 
    stock_prices = [total_paths[i][-1] for i in range(n)]
    put_prices = [max(K-stock_prices[i], 0) for i in range(n)]
    thetaj = [0]*(m+1)
    thetaj[-1] = K

    # Backwards induction for exercise boundary
    for i in range(m-1, -1, -1):
        put_prices = (np.array(put_prices)*np.exp(-r*T/m)).tolist()
        stock_prices = [row[i] for row in total_paths]
        thetaj_p1 = thetaj[i+1]
        (put_prices, thetaj[i]) = opt_exer_bounds(stock_prices, put_prices, K, n, thetaj_p1)
        
    # for Visualization of Exercise boundary 
    plt.figure(figsize = (16, 8))
    ax = sns.lineplot(x = range(len(thetaj)), y = thetaj)
    plt.title('Optimal Exercise Boundary', fontsize = 20)
    plt.xlabel('Time Intervals', fontsize = 16)
    plt.text(0.5, 50, "Continuation region", fontsize=10)
    plt.text(3.5, 45, "Exercise region", fontsize=10)
    plt.ylabel('Stock Price (€)', fontsize = 16)
    ax.xaxis.set_tick_params(labelsize = 16)
    ax.yaxis.set_tick_params(labelsize = 16)
    
    return np.mean(put_prices)
```

```python
#Price and exercise boundary
price = MC_American_put(r, sigma, S0, K, T, m, n)
print('American Put Option price By using Monte Carlo Simulation: €', np.round(price, 2)) 
```

###  Section Three: Gym Environment For Reinforcement Learning

```python
#gym environment created for american put option 

import numpy as np
import gym
class AmeriOptionEnv(gym.Env):
    def __init__(self): 
        self.S0 = 50.0                                                                     # Stock Price 
        self.K = 52.0                                                                      # Strike price or exercise peice
        self.r = 0.05                                                                      # Interest Rate
        self.sigma = 0.30                                                                  # Volatility
        self.T = 0.25                                                                      # maturity
        self.N = 4                                                                         # quarter

        self.S1 = 0                                                                        # intial St
        self.reward = 0                                                                    #intial Reward
        self.day_step = 0                                                                  # from day 0 taking N steps to day N

        self.action_space = gym.spaces.Discrete(2)                                         # 0: hold, 1:exercise
        self.observation_space = gym.spaces.Box(low=np.array([0, 0]), 
                                            high=np.array([np.inf, 0.25]), dtype=np.float32) # S in [0, inf], tao in [0, 0.25]

    def step(self, action):

        if action == 1:                                                                     # exercise
            reward = max(self.K-self.S1, 0.0) * np.exp(-self.r * self.T * (self.day_step/self.N))
            done = True
        else:                                                                               # hold
            if self.day_step == self.N:                                                     # at maturity
                reward = max(self.K-self.S1, 0.0) * np.exp(-self.r * self.T)
                done = True
            else:                                                                           # move to tomorrow
                reward = 0
                self.S1 = self.S1 * np.exp((self.r - 0.5 * self.sigma**2) * (self.T/self.N) + self.sigma * 
                                           np.sqrt(self.T/self.N) * np.random.normal())
                self.day_step += 1
                done = False

        tao = 0.25-self.day_step/self.N                                                     # time to maturity, in unit of years
        return np.array([self.S1, tao]), reward, done, {}

    def reset(self):
        self.day_step = 0
        self.S1 = self.S0
        tao = 0.25-self.day_step/self.N                                                     # time to maturity, in unit of years
        return [self.S1, tao]
    
    def render(self):                                                                       # Visulazing 
        pass

    def close(self):                                                                        # Complete environment or learning 
        pass

env_name = AmeriOptionEnv()
```

## Section Four: Reinforcement Learning

```python
# Import Required Libraries

import gym 
import numpy as np
from collections import namedtuple
from itertools import count

import torch
import torch.nn as nn
import torch.optim as optim   
import torch.nn.functional as F
from torch.distributions import Categorical
```

```python
# Checking gym version
# gym version 0.25.2 required

if not gym.__version__ =='0.25.2':
  raise TypeError("""WARNING! gym version '0.25.2' required..
                    To install gym, do:
                    pip install gym==0.25.2
                    """)
```

```python
#Building Neural Network

class Policy(nn.Module):
    def __init__(self):
        super(Policy, self).__init__()
        self.layer1 = nn.Linear(2, 128)               # 2 as agent will observe whether should exercise or not
        self.layer2 = nn.Linear(128, 1) 
        self.layer3 = nn.Linear(128, 2)               # 2 for the number of actions
        self.saved_actions = []
        self.rewards = []
        self.action = 0
    def forward(self, x):
        x = F.relu(self.layer1(x))
        state_values = self.layer2(x)
        action_prob = F.softmax(self.layer3(x), dim=-1, dtype=torch.double)
        return action_prob, state_values
```

```python
# Hyper-parameters

batch_size = 128                                                # number of bacth size in iteration
gamma = 0.99                                                    # Discount factor
num_iterations = 20000                                          # number of iteration in one epoch
learning_rate = 1e-3                                            # 0.001
log_interval = 10                                               # print log after 10 iterations
eps = np.finfo(np.float32).eps.item()                           #Epsilon
```

```python
# Optimize Q Network & store experience gather from training

model = Policy()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

SavedAction = namedtuple('SavedAction', ['log_prob', 'value'])
```

```python
# Wrapping american option environmentthe env

env = gym.Wrapper(env_name, new_step_api=True)
```

```python
#Print number of action

print("There are {} actions".format(env.action_space.n))
```

```python
 # calculate the losses and perform backpropgation

def select_action(state):
    state = torch.from_numpy(state).float()
    probs, state_value = model(state)
    m = Categorical(probs)
    action = m.sample()
    model.saved_actions.append(SavedAction(m.log_prob(action), state_value))
    return model.action

def episodeEnd():
   
    R = 0
    saved_actions = model.saved_actions
    policy_losses = []
    value_losses =[]
    returns = []
    
    for r in model.rewards[::-1]:
        R = r + 0.99 * R # 0.99 is gamma number                               #Bellman Equation
        returns.insert(0, R)
    returns = torch.tensor(returns)
    returns = (returns - returns.mean()) / (returns.std() + eps)              #Normalization
    
    for (log_prob, value), R in zip(saved_actions, returns):                  # Combine two lists
        advantage = R - value.item()
        
        policy_losses.append(-log_prob * advantage)                           # Policy loss 
        value_losses.append(F.smooth_l1_loss(value, torch.tensor([R])))       # Value Loss 
    
    optimizer.zero_grad()
    loss = torch.stack(policy_losses).sum() + torch.stack(value_losses).sum() #Calculate total loss
    # print("Loss: ", loss.item())
    
    loss.backward()                                                           # Backpropagation
    optimizer.step()                                                          # Update model
                                                                              # Delete previous info
    del model.rewards[:]
    del model.saved_actions[:]
```

```python
# Train Agent

def train(num_episodes = 1000):
  """
  default num_episodes = 1000
  """

  running_reward = 0                                                    #avg reaward of each episode
  for i_episode in range(num_episodes):                                 # iteration over number of episodes
    state = env.reset()                                                 # reset environment
    state = np.array(state, dtype=np.float32)                           # convert into numpy

    ep_reward = 0                                                       # total reward for each episode
    for t in count():
      action = select_action(state) 
      state, reward, terminated, truncated, _ = env.step(action)
      model.rewards.append(reward)
      ep_reward += reward                                                # Each episode reward increment by each action reward
      done = terminated or truncated
      if done:
          break
    running_reward += ep_reward                                          # total episode reward increment by each episode reward
    episodeEnd() 

    if i_episode % log_interval == 0:                                    # print after each log interval
      print("Episode {}\tLast Reward: {:.2f}\tAverage reward: {:.2f}".format(
          i_episode, ep_reward, running_reward/(i_episode+1)   
      ))
```

```python
%%time
train(num_iterations)
```

```python
def compute_avg_return(num_episodes=10):            
  # env_2 = copy.deepcopy(env) # clone of env. Should not alter thr original on reset()
  env_2 = env

  running_reward = 0                                                 #avg reaward of each episode
  total_return = 0.0                                                 # Total return accross all episode
  for i_episode in range(num_episodes):                              # iteration over number of episodes
    state = env_2.reset()                                            # reset environment
    state = np.array(state)                                          # convert into numpy
    ep_reward = 0                                                    #total reward for each episode
    for t in count():
      action = select_action(state)  
      state, reward, terminated, truncated, _ = env_2.step(action)      
      ep_reward += reward                                            # Each episode reward increment by each action reward
      done = terminated or truncated
      if done:
        break

    running_reward += ep_reward                                      # total episode reward increment by each episode reward
    total_return += running_reward/(i_episode+1)                     #avg of total episode reward
  avg_return = total_return / num_episodes
  # print("Avg Reword:\t",avg_return)
  return avg_return
```

## Section Five: European Option Pricing to see early exercise premium

```python
#Import Required Libraries
import numpy as np
from math import exp, sqrt, log
import random
from scipy.stats.distributions import norm
```

```python
#MC Simulation

def monte_carlo_simulation_E(s0, k, r, sig, dt, m, n):
    #list_1 = []                                      # call option value list
    list_2 = []                                       # put option value list
    delta_t = dt / m                                  #length of time interval

    for i in range(n):
        path = [s0]
        for j in range(m):
            path.append(path[-1] * np.exp((r - sig ** 2/2) * delta_t + (sig * np.sqrt(delta_t) * np.random.normal(0, 1))))
        put_value = max(k - path[-1], 0)
        list_2.append(put_value)
    p = np.average(list_2)
    #return (p)
    return np.exp(-r * dt) * np.mean(p)
```

```python
# Price European Option
Pricing = monte_carlo_simulation_E(50, 52, 0.05, 0.3, 0.25, 4, 1000)
print('European Put Option price By using Monte Carlo Simulation: €', np.round(Pricing, 2)) 
```
