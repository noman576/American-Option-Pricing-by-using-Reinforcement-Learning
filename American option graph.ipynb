### 1.3.1 Call Option Graph

```python
op.single_plotter(spot=106, strike=100, op_type='c',tr_type='b', op_pr=5)
```

### 1.3.2 Put Option Graph

### Call Option for buy and sell Graph

```python
op_1 = {'op_type':'c','strike':100,'tr_type':'b','op_pr':5}
op_2 = {'op_type':'c','strike':100,'tr_type':'s','op_pr':5}
op.multi_plotter(spot=106, op_list=[op_1,op_2])
```
